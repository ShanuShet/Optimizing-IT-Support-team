```python

%pip install openpyxl

import pandas as pd
df = pd.read_excel(r"C:\Users\priya\infosys\Optimizing-IT-Support\data\CustomerCallList.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip()
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("/._")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("-","")
df["Phone_Number"] = df["Phone_Number"].str.replace("/","")
df["Phone_Number"] = df["Phone_Number"].str.replace("|","")
df
```

```python
df = df.drop(columns=["Not_Useful_Column"])
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No","N")
df["Paying Customer"] = df["Paying Customer"].str.replace("N/a","N")
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("N/a","N")
df
```

```python
df = df[df["Do_Not_Contact"] != "Y"]
df
```

```python
df = df.fillna(" ")
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace(" ","N")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("Na"," ")
df
```

```python
df = df[df["Phone_Number"] != " "]
df
```

```python
df = df.reset_index(drop=True)
df
```
